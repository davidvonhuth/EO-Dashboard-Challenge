```python
import pandas as pd
import helpers as helpers
import matplotlib.pyplot as plt
import numpy as np

PATH_TO_DATA = "./data/6348800_Q_Month.txt"

df = helpers.parse_discharge_data(PATH_TO_DATA)
```

```python
# plot

def plot_average_against_year(_df, _year):
    # Plotting average for all years (2010-2019)
    averages = helpers.get_means_by_month(_df)
    selected_year = _df[_df['timestamp'].map(lambda x: x.year) == _year]['discharge'].to_numpy()
    months = range(1,13)
    plt.plot(months, averages, label='Average water discharge 1918-1998')
    plt.plot(months, selected_year, label=f'Water Discharge {_year}')
    plt.legend()
    plt.title('Average Monthly Water Discharge levels (mÂ³/s) for river PO, Italy')
    plt.show()
    
    deviations = selected_year - averages
    plt.title(f"deviations for {_year}")
    plt.bar(months, deviations)
    plt.show()
    
    devs_abs = abs(deviations)
    norm_dev = (devs_abs-devs_abs.min())/(devs_abs.max()-devs_abs.min())
    plt.title(f'Normalized deviations for {_year}')
    plt.bar(months, norm_dev)
    plt.show()
```

```python
plot_average_against_year(df, 1995)
```
