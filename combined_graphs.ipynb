### Installing packages, setting up variables

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
devs = [-0.008, -0.408, -0.65, 0.413]
factor = ['Wind', 'Rain', 'Water discharge', 'SST']

fig = plt.figure(figsize=(3,6))
ax = fig.add_axes([0,0,1,1])

width = [0.7,0.7,0.7, 0.7]
x_pos = [1,1.8,2.6,3.4]

ax.bar(x_pos, devs, width=width)
plt.xticks(x_pos, factor, fontsize=7)
plt.yticks(fontsize=7)

# plt.ylim(-1, 1)
plt.title('Normalized deviation from normal levels')
plt.ylabel('<-- less than normal              more than normal -->')
plt.xlabel('factor')

plt.hlines(0, 0.5, 3.9)

for i, value in enumerate(devs):
    if value > 0:        
        plt.text(x_pos[i]-0.1, value+0.01, str(np.round(value,2)), fontsize=7)
    else:
        plt.text(x_pos[i]-0.1, value-0.02, str(np.round(value,2)), fontsize=7)
    
plt.show()
```
